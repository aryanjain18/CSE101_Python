```python
#ENCRYPTION
'''You are given a string. You need to encrypt the string using the following rules →

Convert the string in lowercase characters only
If the character presents as ith position in the English alphabet series, then replace it with (i+n)th character, consider the English alphabet series as circular. Note, if the changed character comes to be a vowel, then follow rule 3 to convert that into the corresponding integer.
If the character is a vowel, then make that a numeric character according to the vowel occurrence in the English alphabet, like 1 for a, 2 for e, and 5 for u.
Add ‘@’ at the starting and ending of the encrypted string.'''

import string
l = list(string.ascii_lowercase)
list_l = l*10
n = int(input())
y = list(input().lower())
ans = []
def encryption(x):
    
    if x in list_l:
        r = list_l.index(x)
        r = list_l[r+n]
        if r in "aeiou":
            return [0,"a","e","i","o","u"].index(r)
        else:
            return r
    else:
        return x
    
for element in y:
    ans.append(encryption(element))

string = ""
for x in ans:
    string+=str(x)
print("@"+string+"@")
```

```python
n = int(input("Enter Number Of Files - "))

import random

def cleanword(string):
    ans = ""
    for i in range(len(string)):
        if string[i].isalpha():
            ans += string[i]
    return ans

def consvchar(string):
    flag = 0
    for elem in ["..",".,",".:",".;",",.",":.",";.",",:",",;",":,",";,",":;",";:"]:
        if elem in string:
            flag = 1
            break
    if flag == 1:
        return True
    else:
        return False

for k in range(1,n+1):
    list_sentences = []
    words = {}
    sen_count = 0
    wor_count = 0
    f3_count = 0
    f4_count = 0

    f = open("Files/Assign3_Q3.%s.txt" %k,"r")
    tempstr = f.readline().strip() #Loop To Read Data
    while tempstr != "":
        if ". " in tempstr:
            lis_sent = tempstr.split(". ")
            for elem in lis_sent:
                list_sentences.append(elem)
        else:
            list_sentences.append(tempstr)
        tempstr = f.readline().strip()
    sen_count = len(list_sentences)
    for i in range(sen_count):
        string = list_sentences[i]
        wordslist = string.split()

        if len(wordslist) > 35 or len(wordslist) < 5:
            f3_count += 1
        for elem in wordslist:
            if consvchar(elem) == True:
                f4_count += 1
            if cleanword(elem.lower()) not in words:
                words[cleanword(elem.lower())] = 0
            words[cleanword(elem.lower())] +=1
    for y in words.values():
        wor_count += y
    lis_words = []
    for x,y in words.items():
        lis_words.append((y,x))
    lis_words.sort(reverse=True)
    f1 = len(lis_words)/wor_count
    f2 = (lis_words[0][0]+lis_words[1][0]+lis_words[2][0]+lis_words[3][0]+lis_words[4][0])/wor_count
    f3 = f3_count/sen_count
    f4 = f4_count/wor_count
    if wor_count > 750:
        f5 = 1
    else:
        f5 = 0
    netscore = 4 + (f1*6) + (f2*6) - f3 - f4 - f5
    with open("Files/scores.txt","a") as file:
        file.writelines("Data corresponding to file - Assign3_Q3.%s.txt \n" %k)
        file.writelines("Score - %s \n" %netscore)
        for j in range(5):
            file.writelines(lis_words[j][1])
            file.writelines(" ")
        file.writelines("\n")
        random.shuffle(lis_words)
        for j in range(5):
            file.writelines(lis_words[j][1])
            file.writelines(" ")
        file.writelines("\n")
        file.writelines("\n")
```
