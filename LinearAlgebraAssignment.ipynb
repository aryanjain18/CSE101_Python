### SOLVE AX = B IN PARAMETRIC FORM

```python
# USING MODULE

from sympy import *
m = int(input("Enter number of rows - "))
print()
f=open("Files/Matrix.txt","r")
matrix=[]

for i in range(m):
    n=list(map(float,f.readline().split()))
    matrix.append(n)

n = len(matrix[0])
M = Matrix(matrix)    
rref = list(M.rref()[0])

lis=[]
i=0
while i<len(rref):
    lis.append(rref[i:i+n])
    i=i+n
print("RREF is ")

for i in lis:
    for j in i:
        print(j,end=" ")
    print()

pivots = []
for i in range(m):
    for j in range(n):
        if lis[i][j]!=0:
            pivots.append((i,j))
            break

print()
print("Pivot Indices - ")
for i in pivots:
    print(i)

tr2 = [[lis[j][i] for j in range(len(lis))] for i in range(len(lis[0]))]
freecol=[]
piv_col=[pivots[i][1] for i in range(len(pivots))]

for j in range(n):
    if j not in piv_col:
        freecol.append(j+1)

print()
print("Parametric Form =>")
print("X = ", end = ' ')
for i in freecol:  
    p=[]
    for j in range(m):
            p.append(0-(lis[j][i-1]))
    for f in freecol:
        if f==i:
            p.insert(i-1, 1)
        else:
            p.insert(f-1, 0)
    print("+", end = '')
    print("x"+str(i)+" *",p[:n],end="  ")
```
